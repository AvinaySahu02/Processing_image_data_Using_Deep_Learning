Converting RGB image to Grayscale image using OpenCV

```python
#importing OpenCV library CV stands for computer vision
import cv2
```

```python
img = cv2.imread('/content/drive/MyDrive/Deeep Learning Project/DL Projects/dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```
